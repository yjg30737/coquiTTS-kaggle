# Import necessary packages

```python
import os
import torch
from TTS.api import TTS
```

# Variables & Making output directory

```python
default_path = 'output'
# Get device
device = "cuda" if torch.cuda.is_available() else "cpu"

speaker_wav = "../input/voice-of-serana/serana.mp3"
os.makedirs(default_path, exist_ok=True)
```

```python
def run(text, filename):
    filename = os.path.join(default_path, f'{filename}.mp3')
    tts.tts_to_file(text=text, speaker_wav=speaker_wav, language="en", file_path=filename)
```

# Load model

```python
# Init TTS
tts = TTS("tts_models/multilingual/multi-dataset/xtts_v2").to(device)
```

# Run

```python
# Run TTS
# ‚ùó Since this model is multi-lingual voice cloning model, we must set the target speaker_wav and language
# Text to speech list of amplitude values as output
# Text to speech to a file

content = '''
If Janus, one of Saturn's moons, were to fall towards Earth, the outcome would depend on a variety of factors, including the moon's composition, speed, angle of entry into the Earth's atmosphere, and the Earth's gravitational influence.
Janus is a small moon with a mean radius of about 89.5 kilometers. If it were to enter the Earth's atmosphere, its size would mean that it wouldn't burn up like a small meteorite. Instead, its entry would likely cause immense atmospheric disruption.
The most significant threat would be if Janus made it through the atmosphere and impacted the Earth. Given its size, an impact would be catastrophic, potentially causing global climatic changes and mass extinctions. The energy released would be several magnitudes greater than the largest nuclear detonations.
Even before impact, the gravitational pull of such a large object would have severe effects on Earth. It could cause massive tidal waves, volcanic activity, and earthquakes.
Depending on various factors like its internal structure and the angle of entry, Janus could fragment before impact. However, even fragmented, the pieces would still cause widespread devastation.
While a small part of its mass might burn due to friction with the Earth's atmosphere, this would be insignificant compared to the overall mass of Janus.
Realistically, for Janus to fall towards Earth, it would require an enormous alteration in its current orbit around Saturn, which is highly improbable under normal celestial mechanics.
In summary, if Janus were to fall to Earth, it wouldn't simply burn up in the atmosphere. The consequences would be globally catastrophic, potentially leading to mass extinctions and significant alterations to the Earth's surface and climate.
'''

# run(content, 'output.mp3')

c_arr = list(filter(lambda x: x, content.split('\n')))
for i, c in enumerate(c_arr):
    run(c, i)
```

# Download outputs

```python
import os
import zipfile

sound_files = [os.path.join(default_path, filename) for filename in os.listdir(default_path)]

# Name of the output compressed file
zip_filename = "sounds.zip"

# Create a new ZIP file
with zipfile.ZipFile(zip_filename, "w") as zipf:
    for sound_file in sound_files:
        zipf.write(sound_file)
        
# Name of the upscale compressed file
        
from IPython.display import FileLink
FileLink(zip_filename)
```

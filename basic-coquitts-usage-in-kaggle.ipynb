```python
import os

import torch
from TTS.api import TTS
# Get device
device = "cuda" if torch.cuda.is_available() else "cpu"

# Init TTS
tts = TTS("tts_models/multilingual/multi-dataset/xtts_v2").to(device)


tts_manager = TTS().list_models()

all_models = tts_manager.list_models()
print(all_models)

# Run TTS
# ‚ùó Since this model is multi-lingual voice cloning model, we must set the target speaker_wav and language
# Text to speech list of amplitude values as output
# Text to speech to a file
tts.tts_to_file(text='If Janus, one of Saturn\'s moons, were to fall towards Earth, the outcome would depend on a variety of factors, including the moon\'s composition, speed, angle of entry into the Earth\'s atmosphere, and the Earth\'s gravitational influence.', speaker_wav="../input/voice-of-serana/serana.mp3", language="en", file_path="output.wav")
```
